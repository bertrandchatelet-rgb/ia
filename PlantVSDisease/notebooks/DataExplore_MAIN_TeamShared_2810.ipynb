```python
# ===== CELL 14 : Debug imports + DataFrame with images =====

import numpy as np
import pandas as pd


from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dropout 
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv2D 
from tensorflow.keras.layers import MaxPooling2D
from keras.utils import to_categorical

from sklearn import metrics

import matplotlib.pyplot as plt
from matplotlib import cm

import itertools

from PIL import Image 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import os

# Fonctions utilitaires
from plant_utils import *


image_path = manage_CWD_arg("Images1")+"\\"
#print(image_path)
df=raw_data_with_images_to_dataframe(image_path)
```

```python
# ===== CELL 15 : Label Encoding =====

from sklearn.preprocessing import LabelEncoder
#print (df.columns) 
#print (df.head())
#y=df["label_all"]
y=df["label_disease"]
if df["label_disease"] != "healthy":
   df["label_disease"]="None-healthy"
#print (y.unique())
#print (df["label_all"].unique())
print (y.value_counts())
X=df["image"]
encoder =  LabelEncoder()
y = encoder.fit_transform(y)
#Quelques verifs
print (y.shape,X.shape)
print (np.bincount(y))
print (sum(np.bincount(y)))
print (np.unique(y))
X_first = np.array(X)
print (y.shape,X_first.shape)
print (X_first[0].shape)
#print (X_first[0])
#X=np.reshape(X_first,(X_fisrt.shape[0],256,256,3))
print (X_first.shape)
#X = np.zeros((X_first.shape[0],256,256,3), dtype='uint8')
#print (X.shape)
#for i in range(X_first.shape[0]):
#    X[i,:,:,:]=X_first[i] 
X=np.array(X_first.tolist())
#print(X[0,0:1,0:1,:])  
for i in range(X.shape[0]):
    X[i,:,:,0:4]=X[i,:,:,0:4]/255
#print(X[17571,0:1,0:1,:]) 
print ('Dimensions: ',X.shape)
```

```python
# ===== CELL 16 : Train / Test split =====

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state=66)
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
print (y_train.shape)
print (y_test.shape)
```

```python
# ===== CELL 17 : Model Architecture =====

inputs = Input(shape = (256, 256,3), name = "Input")

first_layer = Conv2D(filters = 32,
                     kernel_size = (5, 5),
                     padding = 'valid',
                     activation = 'relu')

second_layer = MaxPooling2D(pool_size = (2, 2))



x=first_layer(inputs)
x=second_layer(x)

third_layer = Dropout(rate = 0.2)

fourth_layer = Flatten()

fifth_layer = Dense(units = 128,
                    activation = 'relu')

output_layer = Dense(units = 38,
                     activation='softmax')

x=third_layer(x)
x=fourth_layer(x)
x=fifth_layer(x)

outputs=output_layer(x)



model = Model(inputs = inputs, outputs = outputs)
```

```python
# ===== CELL 18 : Model Training =====

model.compile(loss='categorical_crossentropy', # fonction de perte
              optimizer='adam',                # algorithme d'optimisation
              metrics=['accuracy'])            # métrique d'évaluation

training_history = model.fit(X_train, y_train,
                             validation_split = 0.2,
                             epochs = 10,
                             batch_size = 200)
```

```python
# ===== CELL 19 : Model Evaluation =====

train_acc = training_history.history['accuracy']
val_acc = training_history.history['val_accuracy']
test_pred = model.predict(X_test)

test_pred_class = test_pred.argmax(axis = 1)
y_test_class = y_test.argmax(axis = 1)

print(metrics.classification_report(y_test_class, test_pred_class))
```

```python
# ===== CELL 20 : raw_data_with_images_to_dataframe (with image column) =====

def raw_data_with_images_to_dataframe(data_path): 
    print('data_path:',data_path)
    print('\n--------------------\n * RAW DATA TO DF * \n--------------------\n')
    #data_path = fr'{os.getenv("Images")}'         #data_path est déja en argument de la function 
    #output_csv_file = data_path+'\df_plants.csv'   #il faut changer le chemin, ne pas sauvergaer dans train
    # path_to_raw_data = 'data/RawData/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
    # data_path = os.path.normpath((os.path.join(data_path, path_to_raw_data)))
    mysubfolders = os.listdir(data_path)

    data_dico = dict()
    # Raw Data, Kaggle New Plant Diseases Dataset - 256*256 - RGB
    # 5 keys w/ empty list as value
    data_dico['label_all'] = []
    data_dico['label_plant'] = []
    data_dico['label_disease'] = []
    data_dico['filename'] = []
    data_dico['image'] = []
    #data_dico['filepath'] = []
    #data['image_size'] = []        # removed since all images are 256*256*3
    #data['cat_num_image'] = []
    #data['file_size_meg'] = []
    #data['image'] = []
    #data_dico['image_dim'] = []

    dataNA_dico = dict()            # to list NA (Non Applicable) Data/Image
    dataNA_dico['label_all'] = []

    #NA_file_count = 0
    for subdir in mysubfolders:
        # print(subdir)
        if subdir not in ['desktop.ini']:  # folders & files to ignore
            print(
                ' --- now computing ---------------------------------', subdir, '---')
            # path convention depends on system Unix/windows etc... so need to normpath it !
            folder_path = os.path.normpath((os.path.join(data_path, subdir)))
            # print(folder_path)
            for file in os.listdir(folder_path):
                    if file[-4:] in {'.png', '.PNG', '.jpg', '.JPG'}:
                        file_path = os.path.normpath((os.path.join(folder_path, file)))
                        im = cv.imread(file_path)   # cv2.imread() loads images as BGR
                        if (im is None):            # None returend when imread failed !
                            # print(' ------ I don t want this stuff:', file, ' --- ')
                            dataNA_dico['label_all'].append(subdir)
                            #NA_file_count += 1
                        else:
                            # BGR to RGB converting, reminder: numpy.imread() loads them as RGB
                            im = im[:, :, ::-1]
                            # data['image_dim'].append(im.shape)   # removed since all images are 256*256*3
                            # im = cv.resize(im, (128, 128))       # no more applicable
                            data_dico['label_all'].append(subdir)
                            # label_all split into plant
                            data_dico['label_plant'].append(subdir[0: subdir.find('___')])
                            # and label_all split into disease
                            data_dico['label_disease'].append(subdir[subdir.find('___')+3:])
                            data_dico['filename'].append(file)
                            #data_dico['filepath'].append(file_path)
                            data_dico['image'].append(im)             #decision to don't load the images for the tie being
        else:
            print('--- NA -----', subdir, '---')

    # NA image count
    print('\n------------------------\n * data/image removed * \n------------------------\n')
    print('Amout data/image removed from the RAW dataset:', len(dataNA_dico['label_all']))
    count_NA = Counter(dataNA_dico['label_all'])
    print('\nNA files per Label_all count:\n', count_NA)

    # dico into dataframe
    data_df = pd.DataFrame(data_dico)
    print('\nAmout data/image read/computed with success:',len(data_df.label_all))
    print('\ndf features:',data_df.columns)
    #data_df.to_csv(output_csv_file, sep=',', index=False, header=True)    #a mettre dans un autre fonction
    
    print('\n--------------------\n * data_df issued * \n--------------------\n')
    print('does not include image but image_path!')
    
    return data_df
```
